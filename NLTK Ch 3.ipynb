````
# Basic Meta Characters and Regular Expressions
import re
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
````

````
# $ denotes a token end, ^ denotes a beginning
[w for w in wordlist if re.search('ed$', w)]
````

````
# . is a wildcard. To find all 8-letter words as ^..j..t..$: 
[w for w in wordlist if re.search("^..j..t..$", w)]
````

````
# Ranges also work
[w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)]
````

````
[w for w in wordlist if re.search('^[ghijklmno]+$', w)]
````

````
wsj = sorted(set(nltk.corpus.treebank.words()))
[w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)]

[w for w in wsj if re.search('^[A-Z]+\$$', w)]
[w for w in wsj if re.search('^[0-9]{4}$', w)]
[w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]
[w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
print([w for w in wsj if re.search('(ed|ing)$', w)])
````
